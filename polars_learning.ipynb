```python
import polars as pl
```

```python
data = {
    "id" : [1,2,3,4,5],
    "name" : ['A', 'B', 'C', 'D', 'E'],
    "salary" : [100, 200, 300, 400, 500]
}
```

```python
df = pl.DataFrame(data)
df
```

```python
result = df.filter(
    pl.col("salary") > 200
).select(
    [pl.col("name"), pl.col("salary")]
)

print(result)
```

```python
df_new = df.with_columns(
    (pl.col("salary") * 100).alias("promotion")
)

df_new
```

```python
df_logic = df.with_columns(
    pl.when(pl.col("salary") > 300)
    .then(pl.lit("High"))
    .when(pl.col("salary") > 200)
    .then(pl.lit("Medium"))
    .otherwise(pl.lit("Low"))
    .alias("Salary_tier")
)

df_logic
```
